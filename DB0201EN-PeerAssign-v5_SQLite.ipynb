## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
%%sql
SELECT COUNT(ID) FROM CHICAGO_CRIME_DATA;
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
%%sql 
SELECT community_area_name FROM CENSUS_DATA WHERE per_capita_income < 11000;
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
%%sql 
SELECT ccd.Case_Number 
FROM CHICAGO_CRIME_DATA ccd, CHICAGO_PUBLIC_SCHOOLS_DATA cps
WHERE ccd.Community_Area_Number = cps.Community_Area_Number
AND cps.'Elementary, Middle, or High School' IN ('ES', 'MS', 'HS')
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%%sql
SELECT ID, CASE_NUMBER, PRIMARY_TYPE
FROM CHICAGO_CRIME_DATA 
WHERE DESCRIPTION = 'CHILD ABDUCTION/STRANGER';
```

Elementary, Middle, or High School### Problem 5

##### What kinds of crimes were recorded at schools?


```python
%%sql
SELECT PRIMARY_TYPE, DESCRIPTION
FROM CHICAGO_CRIME_DATA 
WHERE LOCATION_DESCRIPTION = "SCHOOL, PUBLIC, GROUNDS"
```

### Problem 6

##### List the average safety score for each type of school.


```python
%%sql 
SELECT "Elementary, Middle, or High School" AS ES_MS_HS, AVG(SAFETY_SCORE) AS AVG_SAFETY_SCORE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY "Elementary, Middle, or High School"
```

```python
%%sql 
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY
FROM CENSUS_DATA
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5
```

### Problem 8

##### Which community area is most crime prone?


```python
%%sql 
SELECT cd.COMMUNITY_AREA_NAME, COUNT(ccd.ID) AS HighestCrime
FROM CHICAGO_CRIME_DATA ccd, CENSUS_DATA cd
WHERE ccd.COMMUNITY_AREA_NUMBER = cd.COMMUNITY_AREA_NUMBER
GROUP BY cd.COMMUNITY_AREA_NAME
ORDER BY HighestCrime DESC LIMIT 3
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql
select Community_Area_Name, Hardship_Index
from CENSUS_DATA
where Hardship_Index = 
    (
        select MAX(Hardship_Index) 
        from CENSUS_DATA 
    )
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql
SELECT cd.Community_Area_Name AS High_Crime_Area, ccd.Community_Area_Number, COUNT(ccd.ID) AS Number_of_Crimes
FROM CHICAGO_CRIME_DATA ccd, CENSUS_DATA cd
WHERE cd.Community_Area_Number = ccd.Community_Area_Number
AND ccd.Community_Area_Number IN
    (
        SELECT Community_Area_NUMBER
        FROM CENSUS_DATA
    )
GROUP BY cd.Community_Area_NAME
ORDER BY COUNT(ccd.ID) DESC LIMIT 1
```
